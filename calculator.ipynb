```python
# Person A: Create this file as "calculator.ipynb"
# Add this cell and run it to verify it works

# Part 1: Addition (Person A)
def add(a, b):
    """Return the sum of a and b.
    >>> add(3, 4)
    7
    >>> add(-1, 1)
    0
    """
    return a + b

# Part 2: Subtraction (Person A)  
def subtract(a, b):
    """Return a minus b.
    >>> subtract(10, 3)
    7
    >>> subtract(5, 5)
    0
    """
    return a - b

# Test Person A's functions
print("Testing Person A's functions:")
print(f"add(3, 4) = {add(3, 4)}")
print(f"subtract(10, 3) = {subtract(10, 3)}")
print("‚úÖ Person A's functions work!")
```

```python
# Person B: FIRST ‚Äî pull Person A's changes in VS Code (Sync Changes button)
# THEN add these cells to the SAME notebook

# Part 3: Multiplication (Person B)
def multiply(a, b):
    """Return a times b.
    >>> multiply(3, 4)
    12
    >>> multiply(0, 100)
    0
    """
    return a * b

# Part 4: Division (Person B)
def divide(a, b):
    """Return a divided by b. Raises ValueError if b is 0.
    >>> divide(10, 2)
    5.0
    >>> divide(7, 2)
    3.5
    """
    if b == 0:
        raise ValueError("Cannot divide by zero!")
    return a / b

# Test Person B's functions
print("Testing Person B's functions:")
print(f"multiply(3, 4) = {multiply(3, 4)}")
print(f"divide(10, 2) = {divide(10, 2)}")
print("‚úÖ Person B's functions work!")
```
